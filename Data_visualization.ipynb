Dataset used for the problem is **District-wise Production under Mash Cultivation in Punjab from 1968 to 2022 (April to March)** from https://data.gov.in/resource/district-wise-production-under-mash-cultivation-punjab-1968-2022-april-march


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Table_4.7_Production_Mash_0.csv",header=0)
df = df.drop(df.index[-1])
```

**Line plot**

```python
District = df['District/Year']
Mash_cul_2022 = df['2022'].tolist()
Mash_cul_2021 = df['2021'].tolist()
Mash_cul_2020 = df['2020'].tolist()
Mash_cul_2019 = df['2019'].tolist()
Mash_cul_2018 = df['2018'].tolist()
plt.figure(figsize=(10,4),dpi=100)
plt.plot(District,Mash_cul_2022,label='Mash_cultivation_2022')
plt.plot(District,Mash_cul_2021,label='Mash_cultivation_2021')
plt.plot(District,Mash_cul_2020,label='Mash_cultivation_2020')
plt.plot(District,Mash_cul_2019,label='Mash_cultivation_2019')
plt.plot(District,Mash_cul_2018,label='Mash_cultivation_2018')
plt.xticks(rotation = 90)
plt.legend(title='Year')
plt.ylabel('Production')
plt.grid(axis='x')
plt.title('District-wise Production under Mash Cultivation in Punjab from 2018 to 2022 (April to March)')
plt.savefig('Plots/Line.jpg',dpi=300, bbox_inches='tight')
```

In this plot, District-wise production under Mash Cultivation in punjab is more at 2022 compared to other years in all district except for SAS Nagar. Where it is very low during that year. Most of the districts have same production for all years(Fazilka,Faridkot,Moga)

**Scatter Plot**

```python
df['Mash_2022'] = df['2022'].rank(method='min')
df['Mash_2021'] = df['2021'].rank(method='min')
df['Mash_2020'] = df['2020'].rank(method='min')
df['Mash_2019'] = df['2019'].rank(method='min')
df['Mash_2018'] = df['2018'].rank(method='min')
data_sort_rank = df.sort_values(by=['Mash_2022','Mash_2021','Mash_2020','Mash_2019','Mash_2018'])
plt.figure(figsize=(10,4),dpi=100)
plt.scatter(data_sort_rank['District/Year'],data_sort_rank['Mash_2022'],label='2022',marker='o')
plt.scatter(data_sort_rank['District/Year'],data_sort_rank['Mash_2021'],label='2021',marker='+')
plt.scatter(data_sort_rank['District/Year'],data_sort_rank['Mash_2020'],label='2020',marker='*')
plt.scatter(data_sort_rank['District/Year'],data_sort_rank['Mash_2019'],label='2019',marker='^')
plt.scatter(data_sort_rank['District/Year'],data_sort_rank['Mash_2018'],label='2018',marker='.')
plt.xticks(rotation = 90)
plt.legend(title='Ranked based on low Production',loc='center left', bbox_to_anchor=(1, 0.5))
plt.ylabel('Production')
plt.grid()
plt.title('District/Year wise lowest production')
plt.savefig('Plots/Scatter.jpg',dpi=300, bbox_inches='tight')
```

We can see from above plot that slmost all district have same production for all years.

**Box Plot**

Dataset used is **'Sub Divisional Monthly Rainfall from 1901 to 2017'** from 'https://data.gov.in/resource/sub-divisional-monthly-rainfall-1901-2017'

```python
rain_data = pd.read_csv('Sub_Division_IMD_2017.csv')
SI = ['Coastal Andhra Pradesh','Coastal Karnataka','North Interior Karnataka','South Interior Karnataka','Kerala','Tamil Nadu']
rain_data_SI = rain_data[rain_data['SUBDIVISION'].isin(SI)]

plt.figure(figsize=(10,8),dpi=100)
sns.boxplot(x='SUBDIVISION',y='JJAS',data=rain_data_SI,showmeans=True)
plt.savefig('Plots/Box.jpg',dpi=300, bbox_inches='tight')
plt.xticks(rotation=90)
plt.ylabel('Rainfall in mm')
plt.xlabel('')
plt.title('Rainfall in South Indian subdivision for June to Sept. from 1901 to 2017')
plt.savefig('Plots/Box.jpg',dpi=300, bbox_inches='tight')
```

The boxplot shows that the rainfall for the period from June to September for South-Indian subdivisions. Costal Karnataka receives the highest during the period followed by Kerala. The horizontal line represents the median and green triangle represents the mean in figure. Tamil Nadu receives lowest rainfall among the subdivisions for the period.
